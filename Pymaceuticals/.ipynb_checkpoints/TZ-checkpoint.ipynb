```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mdd = pd.read_csv(mouse_drug_data_to_load)
ctd = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset

jdd = pd.merge(mdd,ctd, on='Mouse ID',how = 'outer')

jdd.columns


```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
jdd_org_s = jdd[["Drug","Timepoint","Tumor Volume (mm3)", "Metastatic Sites"]].sort_values(by=["Drug"])
#sort_values(["Drug"])  # .groupby(['Drug'])

#jdd_org_s = jdd_org.groupby(["Timepoint"])

TVM = jdd_org_s["Tumor Volume (mm3)"].mean()


len(jdd_org_s)
type(jdd_org_s)
```

```python
jdd_org_s.count()
```

```python
## An array listing all the drugs
jdd_org_u=jdd_org_s["Drug"].unique()

```

```python
## Creating a new Data Frame containing Drug, Timepoint and TVS
drugs = []
timepoints = []
tvs = []
ms=[]
for x in range (0,len(jdd_org_s)):
    drugs.append(jdd_org_s.iloc[x][0])
    timepoints.append(jdd_org_s.iloc[x][1])
    tvs.append(jdd_org_s.iloc[x][2])
    ms.append(jdd_org_s.iloc[x][3])
    
jdd_org_df = pd.DataFrame({"Drug": drugs, "Timepoint":timepoints, "TVS" :tvs})
jdd_org_dfms = pd.DataFrame({"Drug": drugs, "Timepoint":timepoints,  "MS":ms})
```

```python
drug_dfmj1 = pd.DataFrame(columns = ['Timepoint'])
    #drug = 'Capomulin'
for drug in jdd_org_u:
    ##############################################
    ## Binning Individual Drug DataFrame
    ##############################################
    stdE=[]
    binL=[]
    
    for x in range(len(jdd_org_u)):
        bin_df = jdd_org_s[jdd_org_s["Drug"] == drug]
        bin_LIST = bin_df["Timepoint"].unique()
        binL = bin_LIST.tolist()
        binL.append(-5)
        binL.sort()
        bin_NAMES=[]
        for i in range(1, len(binL)): 
            bin_NAMES.append(str(binL[i]))   
     #bin_LIST = [-5, 0, 5, 10, 15, 20, 25, 30, 35, 40, 45]  
    ###############################################
    Dtime = []
    Dvolume = []
    drug_df = jdd_org_df[jdd_org_df["Drug"] == drug]
    drug_dfms = jdd_org_dfms[jdd_org_dfms["Drug"] == drug] ### Metastatic Sites
    drug_dfs = drug_df.sort_values(['Timepoint'])
    drug_dfmsS = drug_dfms.sort_values(['Timepoint'])
    ###############################################
    ## define sorted dataframe
    ###############################################
    drug_dfg = drug_df.groupby("Timepoint")
    drug_dfgms = drug_dfms.groupby("Timepoint")### Metastatic Sites
    ##drug_dfs.head()
    
    ###############################################
    ## grouping and extracting plot data for mean Cancer Volume
    ###############################################
    drug_df["View Group"] = pd.cut(drug_dfs['Timepoint'], binL, labels = bin_NAMES)
    
    #drug_df["View Group"] = pd.cut(drug_dfs['Timepoint'], bin_LIST)
    #drug_df["View Group"] = pd.cut(drug_dfs['Timepoint'], bins = edges)
    
    drug_dfg1 = drug_df.groupby("View Group")
    drug_dfm1 = drug_dfg1[['Timepoint',"TVS"]].mean()
    drug_dfms1 = drug_dfgms[['Timepoint',"MS"]].mean()
    drug_dfs1 = drug_dfg1[["TVS"]].sem()
    drug_dfm1 = drug_dfm1.rename(columns={"TVS": drug})
    
    
    ##
    ##
    drug_dfmj1 = pd.merge(drug_dfmj1,drug_dfm1, on='Timepoint',how = 'outer')
    
    Dtime = drug_dfm1['Timepoint'].values.tolist()
    Dvolume = drug_dfm1[drug].values.tolist()
    Derror = [value*10 for  value in drug_dfs1['TVS'].values.tolist()]
    Dms = drug_dfms1['MS'].values.tolist()
    stdE.append(drug_dfg.sem())
    
    

    plt.ylim(0,4)
    x_axis = bin_LIST
    plt.title('tumor volume changes over time with ...' + drug)
    plt.xlabel('TIMEPOINTS')
    tick_locations = [value for value in x_axis]
    plt.xticks(tick_locations, bin_LIST)
    plt.ylabel('Average Metastatic Sites')
    plt.scatter(Dtime, Dms,marker="o", facecolors="red", edgecolors="black")
    plt.savefig("../images/"+drug +"_treatmennt")
    plt.show()
        
   
    
#print(Dtime)
```

```python
for drug in jdd_org_u:
    #drug_dfm1 = drug_dfg1[['Timepoint',"TVS"]].mean()
    
    Dvolume = drug_dfmj1[drug].values.tolist()
    Derror = [value*10 for  value in drug_dfs1['TVS'].values.tolist()]
    fig, ax = plt.subplots()
    ax.errorbar(Dtime, Dvolume, Derror, marker='o', mfc='red',  mec='green', ms=5, mew=2 , label = drug)
    ax.set_ylim (0,100)
    ax.set_xlabel('TIMEPOINTS')
    ax.set_ylabel('Tumor Volume [mm3]')
    plt.legend(loc="best", fontsize="small", fancybox=True)
    
plt.show()
```
